## 題目一 (50%)

這是一個整理自國小國文教科書資料集的dataframe，顯示「把」字的不同詞義在國小國文教科書中的使用情況（詞義資訊來自[中文詞彙網路CWN 2.0](https://lopentu.github.io/CwnWeb/)）。這個dataframe的直欄包括：

  - Lemma: 詞條（也就是「把」）
  - Part-of-Speech: 詞類
  - Sense_id: 詞義編號
  - Gloss: 詞義解釋
  - Confidence: 信賴度
  - Grade: 年級
---

## 程式要求
1. 資料讀取
  - 讀取 `ba_df.csv` 檔案，並將其存放在 `ba_df`。
  - 用 `.head(10)` 來預覽資料前10筆記錄（不用印出來）。

2. 資料前處理
  - 將Grade欄位中的同年級統一，如"3上", "3下"合併為"三"。
  - 移除有缺失值的資料，並重新設定索引。
  - 移除在這個任務中不會用到的欄位 `Sense_id` 和 `Confidence`。
  - 將更新後的dataframe用 `.head()` 印出來。

3. 資料觀察
  - 印出「把」在不同年級的使用頻率。
  - 印出「把」的每個詞義。
  - 印出「把」每個詞義的使用頻率。
  - 計算「把」每個詞類出現的次數，並畫成一個柱狀圖。

4. 資料統計
  - 依照年級和詞義分組，計算每個年級使用「把」不同詞義的頻率，並新增欄位為Count。
  - 將更新後的dataframe存放在 `Count_df` 並印出來。

5. 資料視覺化
  - 使用小題4的 `Count_df`，畫出「把」字不同詞義在國小國文教科書中使用分佈的柱狀圖，如下圖。
  - 設定字型為 `Taipei Sans TC Beta` 以確保圖表中的中文能正常顯示。

6. 資料詮釋
  - 增加一個文字cell，並使用Markdown語法簡單描述從圖表觀察到的現象。

## 題目二 (50%)

這是一個來自Dcard的資料集，總共有六個直欄：

  * title: 標題
  * commentCount: 留言數
  * likeCount: 按讚數
  * forumName: 看板名稱
  * gender: 發文者性別
  * text: 內文，格式是list of list of list (以字串資料型態呈現)
    * 第一層是整篇文章
    * 第二層是句子
    * 第三層是 `[token, POS tag]`

---
# 程式要求
1. 讀取資料
  * 讀取 `dcard-500.csv` 檔案，並將其存放在 `dcard_df`。
  * 用 `.head(6)` 來預覽資料前6筆記錄（不用印出來）。

2. 資料前處理
  * 將 `text` 中的token（詞元）和POS tag（詞類標記）分開
    * 將每篇文章中的token和POS tag分別蒐集起來（所有句子放在一起）
    * 使用 `|` 符號將所有的token和POS tag分別連接起來
    * 將token存到 `tokens` 欄位，POS tag存到 `pos` 欄位
    > 如：`[['我', 'Nh'], ['是', 'SHI'], ['一個', 'Neu'], ['好人', 'Na']]`  
          → `tokens = 我|是|一個|好人, pos = Nh|SHI|Neu|Na`
  * 將 `gender` 欄位的數字轉換成文字
     * 0 → `F` 代表Female
     * 1 → `M` 代表Male
   * 刪掉 `text` 欄位
   * 用 `.head()` 將更新後的dataframe印出來
  
3. 資料觀察
   * 印出每個看板的文章數量
   * 印出每個看板的留言數中位數
   * 印出每個看板的按讚數中位數
   * 印出每個看板的發文者性別數量
